```python
# Import Dependencies
import psycopg2
import pandas as pd
from sqlalchemy import create_engine

# Import PostgreSQL password
from config import db_password
```

```python
# Create engine instance
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/carbon_db"
engine = create_engine(db_string)
```

```python
# Connect to PostgreSQL server
dbConnection = engine.connect()
```

```python
# Read data from historical_emissions table into a DataFrame
emissions_df = pd.read_sql("select * from \"historical_emissions\"", dbConnection)

# Display Dataframe
emissions_df.head(10)
```

```python
# Close the database connection
dbConnection.close();
```
